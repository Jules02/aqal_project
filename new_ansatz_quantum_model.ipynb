```python
import networkx as nx
import matplotlib.pyplot as plt
import random
import pennylane as qp
from pennylane import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.metrics import confusion_matrix, ConfusionMatrixDisplay
from IPython.display import clear_output
import pandas as pd
import json
from datetime import datetime, time
import time
```

Note: see the project instructions for notations ($L$, $H_G$, etc.)

```python
num_qubits = num_nodes = 8
L = 2
N_PARAMS_PER_LAYER = 2

DATASET_SIZE = 1000
TEST_SIZE = 0.2

EPOCHS = 50
STEP_SIZE = 0.1
SEED = 1010
random.seed(SEED)
np.random.seed(SEED)

dev = qp.device("default.qubit", wires=num_qubits)
```

```python
config = {
    "experiment_name": "epochs_new_ansatz_filtered",
    "num_nodes": num_nodes,
    "num_layers": L,
    "num_params_per_layer": N_PARAMS_PER_LAYER,
    "parameterized_encoding": True,
    "dataset_size": DATASET_SIZE,
    "test_size": TEST_SIZE,
    "step_size": STEP_SIZE,
    "epochs": EPOCHS,
    "seed": SEED
}
```

### Parametrized Quantum Circuit

```python
def data_encoding_circuit_block(ising_hamiltonian, gamma):
    qp.exp(ising_hamiltonian, coeff=gamma*1j)

def variational_layer(theta, phi):
    for i in range(num_qubits):
        qp.RX(theta[0], wires=i)
        if len(theta) >= 2:
            qp.RY(theta[1], wires=i)
        if len(theta) >= 3:
            qp.RZ(theta[2], wires=i)
    for i in range(num_qubits):
        j = (i + 1) % num_qubits
        qp.IsingXX(phi, wires=[i, j])

@qp.qnode(dev)
def pqc(theta_params, phi_params, gamma_params, H_G):
    for i in range(num_qubits):
        qp.Hadamard(wires=i)
    for j in range(L):
        data_encoding_circuit_block(H_G, gamma_params[j])    # M
        variational_layer(theta_params[j], phi_params[j])     # W_j

    return qp.probs(wires=range(num_qubits))
```

### Metrics

```python
def bce_loss(theta_params, phi_params, gamma_params, H_G, y):
    """'Computes the BCE loss between the prediction of the PQC (given the current theta_params and the graph encoding H_G) and the true label"""
    probs = pqc(theta_params, phi_params, gamma_params, H_G)
    prediction = probs[0]                             # h_k(M)

    eps = 1e-15
    pred_clipped = qp.math.clip(prediction, eps, 1.0 - eps)

    return -(y * qp.math.log(pred_clipped) + (1 - y) * qp.math.log(1 - pred_clipped))

def cost_function(theta_params, phi_params, gamma_params, dataset=None):
    """Computes the mean of all BCE losses over a given dataset"""
    losses = []

    for _, H_G, y in dataset:
        loss_value = bce_loss(theta_params, phi_params, gamma_params, H_G, y)
        losses.append(loss_value)

    return qp.math.mean(qp.math.stack(losses))

def compute_accuracy(theta_params, phi_params, gamma_params, dataset=None):
    """Computes the accuracy of the PQC (proportion of correct predictions) on a given dataset"""
    correct_predictions = 0

    for _, H_G, y in dataset:
        probs = pqc(theta_params, phi_params, gamma_params, H_G)
        prediction = probs[0]

        discrete_pred = 1 if prediction >= 0.5 else 0
        if discrete_pred == y:
            correct_predictions += 1

    return correct_predictions / len(dataset)
```

### Utils

```python
def get_ising_hamiltonian(adj_matrix):
    """
    Returns the Ising Hamiltonian for a given adjacency matrix, as defined in the project instructions.
    If the graph has no edges, returns the zero Hamiltonian.
    """
    n = adj_matrix.shape[0]
    coeffs = []
    obs = []
    for i in range(n):
        for j in range(i + 1, n):
            if adj_matrix[i, j] == 1:
                coeffs.append(1.)
                obs.append(qp.PauliZ(i) @ qp.PauliZ(j))
    if len(obs) == 0:
        return qp.Hamiltonian([0.0], [qp.Identity(0)])
    return qp.Hamiltonian(coeffs, obs)
```

### Data Generation

```python
dataset = []

threshold = np.log(num_nodes) / num_nodes   # The theoretical threshold for a graph to be connected is ln(n)/n
edge_probabilities = [random.uniform(threshold*0.9, threshold*1.5) for _ in range(DATASET_SIZE)]

connected_count = 0
disconnected_count = 0
for p in edge_probabilities:
    G = nx.erdos_renyi_graph(n=num_nodes, p=p)
    A = nx.to_numpy_array(G)
    H_G = get_ising_hamiltonian(A)
    if nx.is_connected(G):
        y = 0
        connected_count+=1
    else:
        y = 1
        disconnected_count+=1
    dataset.append((A, H_G, y))
config["connected_ratio"] = connected_count / len(dataset)

print(f"The full dataset contains {connected_count} connected graphs and {disconnected_count} disconnected graphs.")
train_data, test_data = train_test_split(dataset, test_size=TEST_SIZE, random_state=SEED)
```

```python
def filter_isomorphisms(source_dataset):
    """
    Filters a dataset to contain only non-isomorphic graphs.
    If reference_graphs is provided, it also ensures no graph in the
    source_dataset is isomorphic to any graph in the reference_graphs.
    """

    unique_graphs = list()
    filtered_dataset = []

    for data_point in source_dataset:
        A, H_G, y = data_point
        G_current = nx.from_numpy_array(A)

        is_iso = False
        for G_seen in unique_graphs:
            if nx.is_isomorphic(G_current, G_seen):
                is_iso = True
                break

        if not is_iso:
            unique_graphs.append(G_current)
            filtered_dataset.append(data_point)

    return filtered_dataset

print(f"Original train size: {len(train_data)}")
train_data_clean = filter_isomorphisms(train_data)
print(f"Cleaned train size (no internal isomorphisms): {len(train_data_clean)}\n")

train_data = train_data_clean
```

### Training

```python
thetas = np.random.uniform(0, 2*np.pi, size=(L, N_PARAMS_PER_LAYER), requires_grad=True)
phis = np.random.uniform(-0.1, 0.1, size=(L, ), requires_grad=True)
gammas = np.random.uniform(0, 1, size=(L, ), requires_grad=True)

opt = qp.AdamOptimizer(stepsize=STEP_SIZE)

history_train_loss = []
history_test_loss = []
history_train_acc = []
history_test_acc = []

print("Starting Training...")

total_start_time = time.time()
total_end_time = None
epoch_times = []

last_epoch = None
for epoch in range(EPOCHS):
    epoch_start_time = time.time()

    cost_function_train = lambda t, p, g: cost_function(t, p, g, dataset=train_data)
    thetas, phis, gammas = opt.step(cost_function_train, thetas, phis, gammas)

    train_loss = cost_function(thetas, phis, gammas, dataset=train_data)
    train_acc = compute_accuracy(thetas, phis, gammas, dataset=train_data)
    test_loss = cost_function(thetas, phis, gammas, dataset=test_data)
    test_acc = compute_accuracy(thetas, phis, gammas, dataset=test_data)

    history_train_loss.append(train_loss)
    history_test_loss.append(test_loss)
    history_train_acc.append(train_acc)
    history_test_acc.append(test_acc)

    clear_output(wait=True)

    plt.figure(figsize=(12, 5))

    plt.subplot(1, 2, 1)
    plt.plot(range(1, epoch + 2), history_train_loss, label='Train loss', marker='o', color='blue')
    plt.plot(range(1, epoch + 2), history_test_loss, label='Test loss', marker='x', linestyle='--', color='orange')
    plt.xlabel('Epoch')
    plt.ylabel('Mean BCE loss')
    plt.xlim(1, EPOCHS)
    plt.legend()
    plt.grid(True, linestyle=':', alpha=0.7)

    plt.subplot(1, 2, 2)
    plt.plot(range(1, epoch + 2), history_train_acc, label='Train accuracy', marker='o', color='green')
    plt.plot(range(1, epoch + 2), history_test_acc, label='Test accuracy', marker='x', linestyle='--', color='red')
    plt.xlabel('Epoch')
    plt.ylabel('Accuracy')
    plt.xlim(1, EPOCHS)
    plt.ylim(0.0, 1.05)
    plt.legend()
    plt.grid(True, linestyle=':', alpha=0.7)

    plt.tight_layout()
    plt.show()

    epoch_end_time = time.time()
    epoch_duration = epoch_end_time - epoch_start_time
    epoch_times.append(epoch_duration)

    print(f"Epoch {epoch + 1:3d} / {EPOCHS} | "
          f"Train Loss: {train_loss:.4f} | Train Acc: {train_acc:.2f} | "
          f"Test Loss: {test_loss:.4f} | Test Acc: {test_acc:.2f} | "
          f"Epoch Time: {epoch_duration:.2f} sec")

    if test_acc > 0.75:
        last_epoch = epoch
        total_end_time = time.time()
        break

if total_end_time is None:
    total_end_time = time.time()
total_duration = total_end_time - total_start_time

print("\nFinal Optimized Parameters (thetas):", thetas)
print("\nFinal Optimized Parameters (phis):", phis)
print("\nFinal Optimized Parameters (gammas):", gammas)
```

```python
import networkx as nx
import numpy as np
import random

def generate_novel_test_set(train_data, num_nodes, num_new_graphs):
    """
    Generates a dataset of non-isomorphic Erdős-Rényi graphs that do not
    share an isomorphism class with any graph in the provided training set.
    """
    # Reconstruct NetworkX graph objects from the training adjacency matrices
    # train_data format is assumed to be: [(A, H_G, y), ...]
    train_graphs = [nx.from_numpy_array(item[0]) for item in train_data]

    test_dataset = []
    test_graphs = [] # Keep track of NetworkX objects for the test set

    # The theoretical threshold for a graph to be connected is ln(n)/n
    threshold = np.log(num_nodes) / num_nodes

    print(f"Generating {num_new_graphs} novel non-isomorphic test graphs...")

    while len(test_dataset) < num_new_graphs:
        # Sample around the phase transition threshold
        p = random.uniform(threshold * 0.9, threshold * 1.5)
        G_cand = nx.erdos_renyi_graph(n=num_nodes, p=p)

        # 1. Check if isomorphic to any graph in the training set
        # We use a generator expression with 'any' for early stopping
        is_iso_train = any(nx.is_isomorphic(G_cand, G_t) for G_t in train_graphs)

        # 2. Check if isomorphic to any graph ALREADY added to our new test set
        # This ensures our test set is entirely diverse
        is_iso_test = any(nx.is_isomorphic(G_cand, G_test) for G_test in test_graphs)

        if not is_iso_train and not is_iso_test:
            # Candidate is completely novel! Calculate required tensors
            A_cand = nx.to_numpy_array(G_cand)
            H_G_cand = get_ising_hamiltonian(A_cand)

            # Determine connectivity label
            y_cand = 0 if nx.is_connected(G_cand) else 1

            test_dataset.append((A_cand, H_G_cand, y_cand))
            test_graphs.append(G_cand)

    print(f"Successfully generated {len(test_dataset)} novel graphs.")
    return test_dataset

# Example usage:
novel_test_data = generate_novel_test_set(train_data, num_nodes, num_new_graphs=100)
novel_test_acc = compute_accuracy(thetas, phis, gammas, dataset=novel_test_data)
print(novel_test_acc)
```
